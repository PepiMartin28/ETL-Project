#In this notebook we transform the data that we extract from the notebook "Extraction"


```python
import pandas as pd
import numpy as np
import csv
```

First read the csv to obtain the dataframe

```python
df = pd.read_csv('put your path here')
```

```python
df.head()
```

```python
df.info()
```

Then, create four new columns:

* goals_scored/90
* shots_on_target/90
* goals_against/90
* shots_against/90



```python
df['goals_scored/90'] = np.round((df['goals_scored'] / df['matches_played']), 1)
```

```python
df['goals_against/90'] = np.round((df['goals_against'] / df['matches_played']), 1)
```

```python
df['shots_on_target/90'] = np.round((df['shots_on_target'] / df['matches_played']), 1)
```

```python
df['shots_against/90'] = np.round((df['shots_against'] / df['matches_played']), 1)
```

After, reorder the columns

```python
df = df[['league_id', 'league_name', 'season', 'id', 'name', 'matches_played', 'wins', 'draws', 'losses', 'season_points',
         'goals_scored', 'goals_scored/90', 'shots_on_target','shots_on_target/90',
         'goals_against', 'goals_against/90', 'shots_against', 'shots_against/90','goal_diff',
         'clean_sheets','yellow_cards', 'red_cards']]
```

```python
df.head()
```

Change the type of the columns "yeloow_cards" and "red_cards"

```python
df['yellow_cards'] = np.array(df['yellow_cards'], dtype=int)
```

```python
df['red_cards'] = np.array(df['red_cards'], dtype=int)
```

```python
df.info()
```

Finally export the dataframe as a csv file

```python
df.to_csv('put your path here', index=False)
```
